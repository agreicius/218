```sage
f=plot(-x-2,x,-5,-1, ticks=[[-4,-3,-2,-1,1,2,3,4],[-6,-5,-4,-3,-2,1,2,3,6]])
g=plot(-(x-1)^2+3,-1,3.99)
C=circle((4,-6),0.1)
P=point((4,-2),color="blue", size=50)
t=text("$f(x)$",(-2.5,2), color="black", fontsize=16)
f+g+C+P+t
```

```sage
f1=plot((x-3)/(x^2-4*x+3),x,-2,2.9, ymin=-5, ymax=5, detect_poles="yes")
f2=plot((x-3)/(x^2-4*x+3),x,3.1,6, ymin=-5, ymax=5, detect_poles="yes")
C=circle((3,1/2),.1)
L=line(((1,-5),(1,5)),linestyle="dotted", color="red")
f1+C+f2+L
```
